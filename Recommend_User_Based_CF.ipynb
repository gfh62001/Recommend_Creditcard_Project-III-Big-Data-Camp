# User_based_協同過濾推薦函數

```python
def CF_Recommend_User_Based(my_card):
    #imput data type:my_card=['上海商銀-小小兵Bello回饋卡','中國信託-ANA聯名卡']
    hold_card_df=pd.DataFrame(columns=df.columns)
    hold_card_df.loc[0]=0
    #搜尋使用者持有卡於資料庫並設成1
    hold_card_df.loc[0,my_card]=1
    #計算cosine相似度
    user_sim_array=cosine_similarity(hold_card_df,df)
    sim_value_sort=pd.Series(user_sim_array[0]).sort_values(ascending = False)
    # df_m2m = pd.DataFrame(cosine_similarity(df))
    top3_sim_user={}
    
    
    for idx,v in sim_value_sort.iteritems():
        if len(top3_sim_user)<3:     #top3_sim_user
            top3_sim_user[v]=idx  
    weighted_card_rec=np.zeros(len(df.columns))
    #相似權重相加
    for i in top3_sim_user:
        weighted_card_rec+= df.loc[top3_sim_user[i]]*i  
    #未持有的卡推薦  
    recommend_card=[]
    weighted_card_rec_sort=weighted_card_rec.sort_values(ascending=False)
    for card,value in weighted_card_rec_sort.iteritems():
            if value >0:
                if hold_card_df[card][0] ==0 and len(recommend_card)<5:
                    recommend_card.append(card)
    if recommend_card:
        return recommend_card
    else:
        return '沒有好推薦的卡'
```
